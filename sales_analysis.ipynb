# Exploratory Data Analysis
 Gender

```python
df.columns
```

```python
ax= sns.countplot(x="Gender",data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(["Gender"], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.show()
```

From above graphs we can see that most of the buyers are females and their purchasing power is greater than men

# Age

```python
df.columns
```

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(data=sales_age,x="Age Group",y="Amount")
plt.show()
```

From above graphs we can see that most of the buyers are of age group between 26-35 years female 

# State

```python
df.columns
```

```python
# total number of columns from top 10 states
sales_state=df.groupby(["State"],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(29)

sns.set(rc={"figure.figsize":(25,8)})
sns.barplot(data=sales_state,x="State",y="Orders")
plt.show()
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(25,8)})
sns.barplot(data=sales_state,x="State",y="Amount")
plt.show()
```

From above graphs we can see that unexpectedly most of the orders and totals sales/amount are from Uttar Pradesh, Maharastra and Karnataka.

# Marital Status

```python
ax=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(15,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(5,4)})
sns.barplot(data=sales_status,x="Marital_Status",y="Amount",hue="Gender")
plt.show()
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# Occupation

```python
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.countplot(data=df,x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ocp=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values("Amount",ascending=False)
sns.barplot(data=sales_ocp,x="Occupation",y="Amount")
plt.show()
```

Feom above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product Category

```python
sns.set(rc={"figure.figsize":(30,8)})
ax=sns.countplot(data=df,x="Product_Category",log=True)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(20)
sns.barplot(data=sales_prod,x="Product_Category",y="Amount")
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(20)
sns.barplot(data=sales,x="Product_ID",y="Orders")
plt.show()
```

# Conclusion
 Married women age group 26-35 years from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.
